```python
tf.random.set_seed(42)

model_5 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape = (28, 28)),
    tf.keras.layers.Dense(4, activation = "relu"),
    tf.keras.layers.Dense(4, activation = "relu"),
    tf.keras.layers.Dense(10, activation = "softmax"),
])

model_5.compile(loss = tf.keras.losses.SparseCategoricalCrossentropy(),
                optimizer = tf.keras.optimizers.Adam(learning_rate = 0.001),
                metrics = ["accuracy"])

hist_5 = model_5.fit(train_data_norm,
                     train_labels,
                     epochs = 20,
                     validation_data = (test_data_norm, test_labels))
```

```python
import itertools
from sklearn.metrics import confusion_matrix

def confusion_matrix1(Y_test, y_preds):
  figsize = (10, 10)

  cm = confusion_matrix(Y_test, tf.round(y_preds))
  cm_norm = cm.astype("float") / cm.sum(axis = 1)[:, np.newaxis]
  n_classes = cm.shape[0]

  fig, ax = plt.subplots(figsize = figsize)

  cax = ax.matshow(cm, cmap = plt.cm.Blues) # Here first cm is sonfusion matirx and the second cm is color map. cmap is also color map
  fig.colorbar(cax)

  classes = False

  if classes:
    labels = classes
  else:
    labels = np.arange(cm.shape[0])

  ax.set(title = "Confusion Mtarix",
         xlabel = "Predicted label",
         ylabel = "true label",
         xticks = np.arange(n_classes),
         yticks = np.arange(n_classes),
         xticklabels = labels,
         yticklabels = labels)

  threshold = (cm.max() + cm.min())/2

  for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
    plt.text(j, i, f"{cm[i, j]} ({cm_norm[i, j]*100:.1f}%)",
    horizontalalignment = "center",
    color = "white" if cm[i, j]>threshold else "black",
    size = 7)
```

```python
import random

def plot_random_image(model, images, true_labels, classes):
  i = random.randint(0, len(images))

  target_image = images[i]
  pred_probs = model.predict(target_image.reshape(1, 28, 28))
  pred_labels = classes[pred_probs.argmax()]
  true_labels = classes[true_labels[i]]

  plt.imshow(target_image, cmap = plt.cm.binary)

  if pred_labels == true_labels:
    color = "green"
  else:
    color = "red"

  plt.xlabel("Pred: {} {:2.0f}%(True: {})".format(pred_labels, 100*tf.reduce_max(pred_probs), true_labels), color = color)
```

```python
plot_random_image(model_5, test_data_norm, test_labels, class_names)
```
